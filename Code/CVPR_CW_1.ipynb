```python
img1 = cv2.imread("./FD/1.png")
img2 = cv2.imread("./FD/2.png")
img1 = cv2.cvtColor(img1, cv2.COLOR_RGB2BGR)
img2 = cv2.cvtColor(img2, cv2.COLOR_RGB2BGR)
```

```python
show_both_images(img1, img2)
```

```python
orb = cv2.SIFT_create(50)

kp_img1, des_img1 = orb.detectAndCompute(img1, None)
kp_img2, des_img2 = orb.detectAndCompute(img2, None)

kp_img1_overlayed = cv2.drawKeypoints(img1, kp_img1, outImage=None, color=(255, 0, 0), flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
kp_img2_overlayed = cv2.drawKeypoints(img2, kp_img2, outImage=None, color=(255, 0, 0), flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

show_both_images(kp_img1_overlayed, kp_img2_overlayed)
```
